```python
import pandas as pd
import numpy as np
```

```python
vietnam=pd.read_csv('C:/2nd_project/data_set/total_each_country/total_vietnam.csv')
#vietnam=pd.read_csv('C:/2nd_project/data_set/total_each_country/total_vietnam.csv')
```

```python
vietnam=vietnam.groupby('BEC').sum()
vietnam=vietnam.reset_index()
vietnam=vietnam.astype('int64')
vietnam=vietnam.astype({'BEC':'str'})
vietnam['BEC']= vietnam['BEC'].str.ljust(6, 'x')
vietnam_c1=vietnam[(vietnam['BEC'].str[0]=='1')]
vietnam_c2=vietnam[(vietnam['BEC'].str[0]=='2')]
vietnam_c3=vietnam[(vietnam['BEC'].str[0]=='3')]
vietnam_c4=vietnam[(vietnam['BEC'].str[0]=='4')]
vietnam_c5=vietnam[(vietnam['BEC'].str[0]=='5')]
vietnam_c6=vietnam[(vietnam['BEC'].str[0]=='6')]
vietnam_c7=vietnam[(vietnam['BEC'].str[0]=='7')]
vietnam_c8=vietnam[(vietnam['BEC'].str[0]=='8')]
vietnam_c9=vietnam[(vietnam['BEC'].str[0]=='9')]
vietnam=pd.concat([vietnam_c1,vietnam_c2,vietnam_c3,vietnam_c4,vietnam_c5,vietnam_c6,vietnam_c7,vietnam_c8])
```

```python
country=[]
for i in range(len(vietnam)):
    country.append('vietnam')
    
vietnam.insert(0,"country",country,True)
```
