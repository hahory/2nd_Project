```python
data_au_raw = pd.read_table('Trade_Map_-_List_of_products_exported_by_Australia.txt', sep = '\t')
```

```python
data_au_raw
```

```python
data_au=data_au_raw.drop([0])
```

```python
data_au.rename(columns={'Code':'HS'}, inplace=True)
```

```python
data_au['HS']=data_au['HS'].astype('int64') #hs_bec와 data_au를 병합하기 위해 데이터 타입 int64로 변경
data_au.info()
```

```python
export_au=pd.merge(hs_bec,data_au) #hs_bec와 data_au를 병합하고 열이름 재설정
```

```python
export_au
```

```python
export_au[export_au['BEC'].isnull()] #BEC가 NaN인 행만 추출
```
