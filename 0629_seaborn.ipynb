```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.relplot(x='total_bill',y='tip', data=tips)
```

```python
#hue 색상
sns.relplot(x='total_bill',y='tip', data=tips, hue='smoker')
```

```python
sns.relplot(x='total_bill',y='tip', data=tips, hue='size')
```

```python
sns.relplot(x='total_bill',y='tip', data=tips, style='smoker', hue='size')
```

```python
#색상 선택(구글에서 seaborn palette 검색하면 여러 색상을 볼 수 있음)
sns.relplot(x='total_bill',y='tip', data=tips, hue='size', palette='Blues')
```

```python
#원크기 조절
sns.relplot(x='total_bill',y='tip', data=tips, size='size', palette='Blues')
```
