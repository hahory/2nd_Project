```python
import pandas as pd
import numpy as np
```

```python
export_gm=pd.read_table('Trade_Map_-_List_of_products_exported_by_Germany.txt', sep='\t')
import_gm=pd.read_table('Trade_Map_-_List_of_products_imported_by_Germany.txt', sep='\t')
export_gm.loc[0, 'Code'] = 0 #각 연도별 총액 자료에 활용하기 위해 drop하지 않고 남겨둠
import_gm.loc[0, 'Code'] = 0 #각 연도별 총액 자료에 활용하기 위해 drop하지 않고 
export_gm.drop([export_gm.columns[12]], axis=1, inplace=True) #'Unnamed: 12'열 삭제
import_gm.drop([import_gm.columns[12]], axis=1, inplace=True) #'Unnamed: 12'열 삭제
export_gm['Code']=export_gm['Code'].astype('int64')
import_gm['Code']=import_gm['Code'].astype('int64')
```

```python
hs17_12_4_bec=pd.read_csv('hs17_12_4_bec.csv')
hs17_12_4_bec
```

```python
amount_gm=pd.merge(export_gm,import_gm,how='outer')
amount_gm.rename(columns={'Code':'HS17'}, inplace=True)#HS코드를 먼저 HS17에 맞춰 merge하기 위해 HS17이라 명명함
amount_gm=pd.merge(amount_gm, hs17_12_4_bec, how='left')
```

```python
amount_gm.columns
```

```python
#컬럼 순서 바꾸기
amount_gm=amount_gm[['HS17','HS12', 'HS4', 'BEC5Code1', 'Product label', 'Exported value in 2012',
       'Exported value in 2013', 'Exported value in 2014',
       'Exported value in 2015', 'Exported value in 2016',
       'Exported value in 2017', 'Exported value in 2018',
       'Exported value in 2019', 'Exported value in 2020',
       'Exported value in 2021', 'Imported value in 2012',
       'Imported value in 2013', 'Imported value in 2014',
       'Imported value in 2015', 'Imported value in 2016',
       'Imported value in 2017', 'Imported value in 2018',
       'Imported value in 2019', 'Imported value in 2020',
       'Imported value in 2021']]
```

```python
amount_gm.isnull().sum()
```

```python
no_amount_gm=amount_gm[(amount_gm['Exported value in 2017']==0) 
                       & (amount_gm['Exported value in 2018']==0) 
                       & (amount_gm['Exported value in 2019']==0) 
                       & (amount_gm['Exported value in 2020']==0) 
                       & (amount_gm['Exported value in 2021']==0)
                       & (amount_gm['Imported value in 2017']==0) 
                       & (amount_gm['Imported value in 2018']==0) 
                       & (amount_gm['Imported value in 2019']==0) 
                       & (amount_gm['Imported value in 2020']==0) 
                       & (amount_gm['Imported value in 2021']==0)].index
amount_gm.drop(no_amount_gm, inplace=True)
amount_gm.loc[amount_gm['HS17'] == 0, 'BEC5Code1'] = 000000
amount_gm.loc[amount_gm['HS17'] == 999999, 'BEC5Code1'] = 999999
amount_gm.loc[amount_gm['HS17'] == 999999, 'HS12'] = 999999
amount_gm.loc[amount_gm['HS17'] == 999999, 'HS4'] = 999999
```
